```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('/Users/yaohaorong/Desktop/Alan-local.csv')
data1.set_index(data1.columns[1], inplace=True) 
data1
```

```python
df1 = pd.DataFrame(data1, columns=data1.columns, index=data1.index)
# 获取第一个 DataFrame 的时间序列并调整起点为0秒
df1_time = pd.to_datetime(df1.index)  # 将索引转换为时间序列
df1_time_adjusted = (df1_time - df1_time[0]).total_seconds()  # 以秒为单位调整时间起点为0

df1
```

```python
data2=pd.read_csv('/Users/yaohaorong/Desktop/Alan-cloud.csv')
data2.set_index(data2.columns[1], inplace=True) 
df2 = pd.DataFrame(data2, columns=data2.columns, index=data2.index)
# 获取第二个 DataFrame 的时间序列并调整起点为0秒
df2_time = pd.to_datetime(df2.index)  # 将索引转换为时间序列
df2_time_adjusted = (df2_time - df2_time[0]).total_seconds()  # 以秒为单位调整时间起点为0
df2
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# 绘制折线图
plt.figure(figsize=(10, 6))

# 绘制第一个 DataFrame 的 "GPU1 temperature"
plt.plot(df1_time_adjusted, df1[' GPU1 temperature    '], label='Local GPU1 Temperature')

# 绘制第二个 DataFrame 的 "GPU1 temperature"
plt.plot(df2_time_adjusted, df2[' GPU1 temperature    '], label='Cloud GPU1 Temperature')

# 设置图形标题和标签
plt.title('GPU1 Temperature Over Time')
plt.xlabel('Time (seconds since start)')
plt.ylabel('Temperature')
plt.legend()
```

```python
# 绘制折线图
plt.figure(figsize=(10, 6))

# 绘制第一个 DataFrame 的 "GPU1 usage"
plt.plot(df1_time_adjusted, df1['GPU1 usage          '], label='Local GPU1 Usage')

# 绘制第二个 DataFrame 的 "GPU1 usage"
plt.plot(df2_time_adjusted, df2['GPU1 usage          '], label='Cloud GPU1 Usage')

# 设置图形标题和标签
plt.title('GPU1 Usage Over Time')
plt.xlabel('Time (seconds since start)')
plt.ylabel('usage')
plt.legend()
```

```python
# 绘制折线图
plt.figure(figsize=(10, 6))

# 绘制第一个 DataFrame 的 "GPU1 memeory usage"
plt.plot(df1_time_adjusted, df1['GPU1 memory usage   '], label='Local GPU1 Memeory Usage')

# 绘制第二个 DataFrame 的 "GPU1 memoery usage"
plt.plot(df2_time_adjusted, df2['GPU1 memory usage   '], label='Cloud GPU1 Memory Usage')

# 设置图形标题和标签
plt.title('GPU1 Memory Usage Over Time')
plt.xlabel('Time (seconds since start)')
plt.ylabel('memory usage')
plt.legend()
```

```python
cpu_temperature = [f'CPU{i} temperature    ' for i in range(1, 9)]  # 生成列名列表，从 'CPU1 temperature' 到 'CPU8 temperature'

# 创建一个 4x2 的图像
fig, axs = plt.subplots(4, 2, figsize=(12, 16))
axs = axs.ravel()  # 将二维数组展平，方便遍历

# 遍历每个 CPU 温度列，在对应的子图中绘制折线图
for i, cpu_col in enumerate(cpu_temperature):
    row = i // 2  # 计算子图所在的行索引
    col = i % 2   # 计算子图所在的列索引
    
    # 绘制 df1 中对应 CPU 温度的折线图
    axs[i].plot(df1_time_adjusted, df1[cpu_col], label='Local Data', color='blue')
    
    # 绘制 df2 中对应 CPU 温度的折线图
    axs[i].plot(df2_time_adjusted, df2[cpu_col], label='Cloud Data', color='orange')
    
    # 设置子图标题和标签
    axs[i].set_title(cpu_col.strip())  # 去除列名中的额外空格并作为子图标题
    axs[i].set_xlabel('Time (seconds since start)')
    axs[i].set_ylabel('Temperature')
    axs[i].legend()  # 显示图例

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()

```

```python
# 选择需要的指标列
interest_columns = [' GPU1 temperature    ', 'GPU1 usage          ', 'GPU1 memory usage   ', 'RAM usage           ', 'Framerate           ']

# 计算本地和云端数据每个指标的最大值、最小值和均值
df1_stats = data1[interest_columns].agg(['max', 'min', 'mean'])
df2_stats = data2[interest_columns].agg(['max', 'min', 'mean'])

# 创建一个包含5个子图的 1x5 布局的大图
fig, axs = plt.subplots(1, 5, figsize=(20, 6))

# 遍历每个感兴趣的指标列，在对应的子图中绘制柱状图对比
for i, col in enumerate(interest_columns):
    # 提取本地和云端数据中对应指标列的最大值、最小值和均值
    local_max = df1_stats.loc['max', col]
    local_min = df1_stats.loc['min', col]
    local_mean = df1_stats.loc['mean', col]
    
    cloud_max = df2_stats.loc['max', col]
    cloud_min = df2_stats.loc['min', col]
    cloud_mean = df2_stats.loc['mean', col]
    
    # 绘制柱状图
    bar_width = 0.35
    bar_positions_local = np.arange(3)
    bar_positions_cloud = np.arange(3) + bar_width
    
    axs[i].bar(bar_positions_local, [local_min, local_max, local_mean], width=bar_width, label='Local', color='blue')
    axs[i].bar(bar_positions_cloud, [cloud_min, cloud_max, cloud_mean], width=bar_width, label='Cloud', color='orange')
    
    # 设置子图标题和标签
    axs[i].set_title(col)  # 使用指标名称作为子图标题
    axs[i].set_xticks(bar_positions_local + bar_width / 2)
    axs[i].set_xticklabels(['Min', 'Max', 'Mean'])
    axs[i].set_ylabel('Value')
    axs[i].legend()  # 显示图例

# 调整子图间距和整体布局
plt.tight_layout()
plt.show()
```

```python
# 生成列名列表，从 'CPU1 temperature' 到 'CPU8 temperature'
cpu_temperature = [f'CPU{i} temperature    ' for i in range(1, 9)]

# 创建一个包含8个子图的 4x2 布局的大图
fig, axs = plt.subplots(4, 2, figsize=(12, 16))
axs = axs.ravel()  # 将二维数组展平成一维，方便索引

# 遍历每个 CPU 温度列，在对应的子图中绘制柱状图
for i, cpu_col in enumerate(cpu_temperature):
    row = i // 2  # 计算子图所在的行索引
    col = i % 2   # 计算子图所在的列索引
    
    # 提取本地和云端数据中对应 CPU 温度列的数据
    local_data = df1[cpu_col]
    cloud_data = df2[cpu_col]
    
    # 计算本地和云端数据的最大值、最小值和均值
    local_max = local_data.max()
    local_min = local_data.min()
    local_mean = local_data.mean()
    
    cloud_max = cloud_data.max()
    cloud_min = cloud_data.min()
    cloud_mean = cloud_data.mean()
    
    # 绘制柱状图
    bar_width = 0.2
    bar_positions_local = np.arange(3) - bar_width / 2
    bar_positions_cloud = np.arange(3) + bar_width / 2
    
    axs[i].bar(bar_positions_local, [local_min, local_max, local_mean], width=bar_width, label='Local', color='blue')
    axs[i].bar(bar_positions_cloud, [cloud_min, cloud_max, cloud_mean], width=bar_width, label='Cloud', color='orange')
    
    # 设置子图标题和标签
    axs[i].set_title(cpu_col.strip())  # 去除列名中的额外空格并作为子图标题
    axs[i].set_xticks([0, 1, 2])
    axs[i].set_xticklabels(['Min', 'Max', 'Mean'])
    axs[i].set_ylabel('Temperature')
    axs[i].legend()  # 显示图例

# 调整子图间距和整体布局
plt.tight_layout()
plt.show()
```
